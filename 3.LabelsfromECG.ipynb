```python
import pandas as pd
import ast

# ECG metadata (your 43 patients)
ecg_df = pd.read_csv("/content/ptbxl_database.csv")

# Convert 'scp_codes' column from string to dict
ecg_df['scp_codes'] = ecg_df['scp_codes'].apply(lambda x: ast.literal_eval(x))
```

```python
# List of codes mapped to your labels
AFIB_codes = ['AFIB', 'AFLT']
PVC_codes = ['PVC']
AbSin_codes = ['ABQRS','SARRH','SBRAD','LVH','LPFB','LAFB','RVH','NST_','DIG','LNGQT']
Normal_codes = ['NORM','SR']
```

```python
def get_label(scp_dict):
    codes = scp_dict.keys()
    if any(code in AFIB_codes for code in codes):
        return 'AFib'
    elif any(code in PVC_codes for code in codes):
        return 'PVC'
    elif any(code in AbSin_codes for code in codes):
        return 'AbSin'
    elif any(code in Normal_codes for code in codes):
        return 'Normal'
    else:
        return 'AbSin'  # fallback for unclassified
```

```python
ecg_df['label'] = ecg_df['scp_codes'].apply(get_label)

# Keep only relevant columns
labels_df = ecg_df[['ecg_id', 'patient_id', 'label']]

print(labels_df)
```
